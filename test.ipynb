```python
def valid_move(x, y, grid):
    # Check if the (x, y) position is valid within the grid boundaries
    return 0 <= x < len(grid) and 0 <= y < len(grid[0])

def dfs(x, y, grid, visited):
    # Base case: If the cell is out of bounds or already visited, return 0
    if not valid_move(x, y, grid) or visited[x][y]:
        return 0

    # Mark the current cell as visited
    visited[x][y] = True

    # Start with a count of 1 for the current cell
    count = 1

    # Directions: Right and Down
    directions = [(0, 1), (1, 0)]

    # Explore all valid neighboring cells
    for dx, dy in directions:
        count += dfs(x + dx, y + dy, grid, visited)
    
    print(x,y, count)

    return count

# Example grid (1s are accessible cells, 0s are blocked)
grid = [
    [1, 1, 0],
    [0, 1, 1],
    [1, 1, 1]
]

# Initialize visited matrix to keep track of visited cells
visited = [[False for _ in row] for row in grid]

# Assuming we start from the top-left corner
start_x, start_y = 0, 0

# Perform DFS from the starting cell
reachable_cells = dfs(start_x, start_y, grid, visited)

print(f"Number of reachable cells from ({start_x}, {start_y}): {reachable_cells}")
```

```python
def dfs(node, adj_list, visited, counts):
    """
    Perform DFS from the given node to count all reachable nodes.
    
    :param node: Current node
    :param adj_list: Adjacency list representing the graph
    :param visited: Set of visited nodes in the current DFS path
    :param counts: Dictionary storing the count of reachable nodes for each node
    :return: None
    """
    for neighbor in adj_list[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            counts[node] += 1
            dfs(neighbor, adj_list, visited, counts)
            visited.remove(neighbor)

def count_children_nodes(adj_list):
    """
    Counts the number of child nodes (all reachable nodes) for every node in the graph.
    
    :param adj_list: Adjacency list representing the graph
    :return: A dictionary with the count of reachable nodes for each node
    """
    counts = {node: 0 for node in adj_list}  # Initialize counts for each node to 0
    
    for node in adj_list:
        visited = set([node])  # Initialize visited set with the current node to avoid counting it
        dfs(node, adj_list, visited, counts)
    
    return counts

# Example usage
if __name__ == "__main__":
    adj_list = {
        0: [1, 2],
        1: [2, 3],
        2: [3],
        3: [],
        4: [0, 2, 3]  # Adding another node for complexity
    }

    child_counts = count_children_nodes(adj_list)
    for node, count in child_counts.items():
        print(f"Node {node} can reach {count} {'node' if count == 1 else 'nodes'} directly or indirectly.")
```

```python
def dfs(node, graph, visited):
    """
    Perform a depth-first search (DFS) starting from `node`, marking nodes as visited.

    :param node: The current node being visited.
    :param graph: The adjacency list representing the graph.
    :param visited: A set of already visited nodes to avoid revisiting.
    :return: The number of unique nodes visited during this DFS call, excluding the start node.
    """
    count = 0
    for neighbor in graph[node]:
        if neighbor not in visited:
            visited.add(neighbor)
            count += 1 + dfs(neighbor, graph, visited)
    return count

def count_child_nodes(graph):
    """
    Count the number of child nodes for every node in the graph using DFS.

    :param graph: The adjacency list representing the graph, as a list of lists.
    :return: A dictionary with nodes as keys and the count of child nodes as values.
    """
    child_counts = {}
    for node in range(len(graph)):
        visited = set([node])  # Start with the current node as visited to exclude it from its own count
        child_counts[node] = dfs(node, graph, visited)
    return child_counts

# Example graph represented as an adjacency list (array of arrays)
graph = [
    [1, 2],    # Node 0 is connected to nodes 1 and 2
    [2, 3],    # Node 1 is connected to nodes 2 and 3
    [3],       # Node 2 is connected to node 3
    [],        # Node 3 has no connections
    [0, 2, 3]  # Node 4 is connected to nodes 0, 2, and 3 for a more complex example
]

# Count the child nodes for each node in the graph
child_counts = count_child_nodes(graph)

# Print the results
for node, count in child_counts.items():
    print(f"Node {node} can reach {count} {'node' if count == 1 else 'nodes'} directly or indirectly.")
```
